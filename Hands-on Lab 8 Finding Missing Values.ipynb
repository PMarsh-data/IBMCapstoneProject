##### Import Necessary Modules:

## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
print(df.info())
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
df.isnull()
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
plt.figure(figsize = (20,10))
sns.heatmap(df.isnull(), cbar = False, cmap = "crest", yticklabels = False)
plt.title('Heatmap of missing values')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
missing_rows_Employment = df['Employment'].isnull().sum()

missing_rows_EdLevel = df['EdLevel'].isnull().sum()

print('Number of missing row in Employment column:', missing_rows_Employment)
print('Number of missing row in EdLevel column:', missing_rows_EdLevel)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
most_frequent_Employment = df['Employment'].value_counts().idxmax()

most_frequent_EdLevel = df['EdLevel'].value_counts().idxmax()

print('The most frquent value in Employment is:', most_frequent_Employment)
print('The most frquent value in EdLevel is:', most_frequent_EdLevel)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
df['Employment'] = df['Employment'].fillna(most_frequent_Employment)

df['EdLevel'] = df['EdLevel'].fillna(most_frequent_EdLevel)

missing_rows_Employment_new = df['Employment'].isnull().sum()

missing_rows_EdLevel_new = df['EdLevel'].isnull().sum()

print('Number of missing row in Employment column:', missing_rows_Employment_new)
print('Number of missing row in EdLevel column:', missing_rows_EdLevel_new)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
Employment = df['Employment'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x = Employment.index, y = Employment.values)
plt.xticks(rotation=90)
plt.title("Distribution of Data by Employment")
plt.xlabel("Employment")
plt.ylabel("Count")
plt.show()

EdLevel = df['EdLevel'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x = EdLevel.index, y = EdLevel.values)
plt.xticks(rotation=90)
plt.title("Distribution of Data by EdLevel")
plt.xlabel("EdLevel")
plt.ylabel("Count")
plt.show()
```
